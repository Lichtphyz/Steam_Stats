```python
page[:500]
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page,"lxml")
```

```python
print(soup.prettify())
```

```python
#find all the links

for link in soup.find_all('a')[0:80]: 
    print(link)
```

```python
#Get the https:... addresses out of each link
All_Links = []
for link in soup.find_all('a'):
    All_Links.append(link['href'])
All_Links[0:50]
#Links
```

trim off all the undesireable items from the beginning and end
remove all those than don't start with "'https://steamdb.info/app/" 

```python
import copy

base_address = 'https://steamdb.info/app/'
b = len(base_address)
Links = copy.deepcopy(All_Links)

for link in All_Links:
    print(b,link)
    print(link[0:b] != base_address)
    if link[0:b] != base_address:
        print('removing',link)
        Links.remove(link)
```

```python
import numpy as np

np.savetxt("MultiPlayerGames.csv", Links, delimiter=",", fmt='%s', header="Links")
```
