```python
import numpy as np
import pandas as pd
import requests
from time import sleep
from bs4 import BeautifulSoup
import re
from dateutil.parser import *
import numpy.random
```

```python
filename = 'MultiPlayerGames.csv'

df=pd.read_csv(filename, sep=',',header='infer')
df.values
```

```python
Links = []
for line in df.values:
    Links.append(line[0])
Links[0:5]
```

```python
url = Links[0]
#url = "https://steamdb.info/app/252950/"
response = requests.get(url)
```

```python
html = response.text
```

```python
def pullpage(url):
    """download html of each page, and save to disk using Steam ID number"""
    print("trying page at",url)
    response = requests.get(url)
    htmlstring = response.text
    soup = BeautifulSoup(htmlstring,"lxml")
    
    ID_num = soup.find(class_="table table-bordered table-hover table-dark")\
        .findNext("td").findNext("td").text
    Name = soup.find(itemprop="name").text
    
    filename = "/Users/dlicht/ds/metis/metisgh/projects/projectluther/Data/"+\
                    "SteamGameID"+str(ID_num)+".html"
    file = open(filename,'w') 
    file.write(htmlstring) 
    file.close()
    print("saved",Name,"page to",filename)
```

```python
pullpage(url)
```

```python

for game in Links[1750:]:
    pullpage(game)
    delay = 3*np.random.random()+1
    sleep(delay)
print()
print('Mission Accomplished!')
```
